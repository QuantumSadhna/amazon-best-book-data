## _Preparing, Exploring and Analyzing_

```python

###### importing various libaries to work on
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

###### reading data from csv file
file_name= pd.read_csv("bestsellers with categories.csv")

###### displaying few rows
print("Few rows of the dataset:")
print(file_name.head(10))

###### statistics
print(file_name.describe())

###### count of books per genre
plt.figure(figsize=(6,3))
sns.countplot(x='Genre', data=file_name, palette='husl')
plt.title('Distribution of Books by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=5)
plt.show()

###### top 10 authors
top_authors = file_name['Author'].value_counts().head(10)
plt.figure(figsize=(6, 3))
top_authors.plot(kind='bar', color='pink')
plt.title('Top 10 Authors')
plt.xlabel('Author')
plt.ylabel('Number of Bestsellers')
plt.xticks(rotation=45)
plt.show()

###### Correlation between User Rating and Reviews
plt.figure(figsize=(8, 6))
sns.scatterplot(x='User Rating', y='Reviews', data=file_name, hue='Genre', palette='pastel')
plt.title('Correlation between User Ratings and Reviews by Genre')
plt.xlabel('User Rating')
plt.ylabel('Reviews')
plt.legend(title='Genre')
plt.show()

##### Price Impact - correlation for price and user ratings
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Price', y='User Rating', data=file_name, hue='Genre', palette='Set3')
plt.title('Correlation between Price and User Ratings by Genre')
plt.xlabel('Price')
plt.ylabel('User Rating')
plt.legend(title='Genre')
plt.show()
```
